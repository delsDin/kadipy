# KadiPy Market - Module Logistics

Ce notebook calcule les coûts de transfert (transport, tracasseries, qualité) sur les corridors commerciaux.

---

## 1. Initialisation du Module Logistique

Le module `MarketLogistics` n'est pas qu'un simple calculateur de distance. Il modélise les réalités de la logistique agricole en Afrique de l'Ouest (mauvais état des routes, barrages routiers, pertes post-récolte pendant le trajet, prix fluctuant du carburant).

**Pourquoi ?** Connaître le prix du marché ne suffit pas. Si le maïs coûte 200 XOF à Bohicon et 250 XOF à Malanville, la différence de 50 XOF couvre-t-elle le transport ? C'est le rôle de ce module.

```python
import sys
import os
sys.path.append(os.path.abspath('../../'))

from kadi.market.logistics import MarketLogistics

logistics = MarketLogistics()
print("Module Logistics initialisé !")
```

---
## 2. Calcul de la distance d'un corridor (`get_distance`)

La méthode `get_distance` interroge la base de données interne pour obtenir la distance kilométrique exacte entre deux pôles économiques.

**Explication de la sortie :**
Elle retourne simplement un entier (la distance en kilomètres). Ce chiffre est la brique de base pour tous les calculs financiers qui suivent.

```python
dist = logistics.get_distance('Bohicon', 'Malanville')
print(f"Distance géographique Bohicon -> Malanville : {dist} km")
```

---
## 3. Calcul du coût total de transfert (`calculate_transfer_cost`)

C'est la méthode la plus importante du module. Elle prend la ville de départ, la ville d'arrivée, et le prix actuel du carburant (qui influe grandement sur les tarifs des camionneurs).

**Explication de la sortie :**
Au lieu de donner un prix brut inexpliqué, la méthode retourne un dictionnaire détaillé (transparence totale) :
- `total_cost_cfa` : Le coût total pour transporter **1 tonne** de marchandise sur ce trajet.
- `distance_km` : Rappel de la distance.
- `fuel_price` : Le prix du carburant utilisé pour l'estimation.
- `estimated_tracasseries` : Une estimation des frais informels ou barrages sur ce corridor, typique des réalités de terrain.
- `quality_loss_percent` : Le pourcentage estimé de marchandises qui sera perdu ou dégradé à cause des secousses et de la chaleur pendant le trajet.

```python
cout = logistics.calculate_transfer_cost('Bohicon', 'Malanville', prix_carburant=680.0)

print(f"Coût total estimé : {cout['total_cost_cfa']:.2f} XOF / Tonne")
print("\nDétails de la facture logistique :")
for key, value in cout['details'].items():
    print(f"  - {key} : {value}")
```
