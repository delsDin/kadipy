Ce notebook dÃŠmontre l'utilisation de `MarketPricing` et `WFPDataBridgesClient` pour l'ingestion, la normalisation et la dÃŠtection d'anomalies sur les prix du marchÃŠ.

## 1. Initialisation du SystÃ¨me

Nous commenÃ§ons par instancier les classes principales. Nous crÃŠons d'abord un `client` WFP en forÃ§ant le mode `use_local_mirror=True`. 
Cela signifie : *"Ne cherche pas Ã  te connecter Ã  Internet, utilise le simulateur de donnÃŠes interne"*. 
Ensuite, on injecte ce client dans la classe mÃ¨re `MarketPricing`.

**Pourquoi ?** Cela garantit que notre logique mÃŠtier est prÃĒte et dispose d'une source de donnÃŠes valide pour travailler, sans risquer de planter pour des problÃ¨mes de rÃŠseau ou de clÃŠ API.

```python
import sys
import os
import pandas as pd
sys.path.append(os.path.abspath('../../'))

from kadi.market.data_ingestion import WFPDataBridgesClient
from kadi.market.pricing import MarketPricing

# 1. Initialisation avec le mode hors-ligne forcÃŠ (pour l'exemple sans clÃŠ API)
client = WFPDataBridgesClient(use_local_mirror=True)
pricing = MarketPricing(wfp_client=client)

print("Module Pricing initialisÃŠ avec succÃ¨s !")
```

---
## 2. RÃŠcupÃŠration des donnÃŠes (`fetch_prices`)

La mÃŠthode `fetch_prices` est la porte d'entrÃŠe pour rÃŠcupÃŠrer l'historique des marchÃŠs. Elle demande Ã  la source de donnÃŠes (le WFP ou notre miroir local) de lui livrer les prix du maÃ¯s (`maize`) Ã  Savalou sur les 30 derniers jours (`days_back=30`).

**Explication de la sortie :**
La sortie est un tableau (DataFrame pandas) qui possÃ¨de une structure standardisÃŠe essentielle pour KadiPy :
- `date` : Le moment du relevÃŠ.
- `price` : La valeur numÃŠrique du prix sur le marchÃŠ (ici, des prix gÃŠnÃŠrÃŠs alÃŠatoirement autour d'une fourchette rÃŠaliste par notre miroir local).
- `unit` : Toujours formatÃŠ, ici `XOF/kg` (Francs CFA par kilo).

```python
df_prices = pricing.fetch_prices(crop='maize', market='savalou_market', days_back=30)
display(df_prices.head())
```

---
## 3. Normalisation des UnitÃŠs (`normalize_units`)

Dans le monde agricole, les prix sont rapportÃŠs de maniÃ¨re chaotique : en Tonnes, en sacs de 100kg, ou au kilo. 
La mÃŠthode `normalize_units` prend un prix brut, lit son unitÃŠ d'origine (`unit_orig`), et le ramÃ¨ne mathÃŠmatiquement Ã  une ÃŠchelle standard universelle pour KadiPy (le kilogramme).

**Explication de la sortie :**
Le systÃ¨me reconnaÃŽt le mot "Tonne" et divise automatiquement le prix par 1000. C'est crucial : si notre IA de prÃŠvision mÃŠlangeait des prix au kilo et des prix Ã  la tonne, elle produirait des prÃŠdictions totalement erronÃŠes.

```python
prix_tonne = 250000.0
prix_kg = pricing.normalize_units(prix_tonne, unit_orig='XOF/Tonne', crop='maize')
print(f"{prix_tonne} XOF/Tonne = {prix_kg} XOF/kg")
```

---
## 4. DÃŠtection d'anomalies par Z-Score (`detect_anomalies`)

Les collecteurs de donnÃŠes sur le terrain font parfois des erreurs de frappe (ajouter un 0 de trop). 
La mÃŠthode `detect_anomalies` scanne l'historique complet en utilisant un algorithme statistique appelÃŠ **Z-Score**. Il calcule la moyenne des prix et l'ÃŠcart-type (la volatilitÃŠ). Si un point de donnÃŠe s'ÃŠloigne trop de la norme (gÃŠnÃŠralement 3 fois l'ÃŠcart-type), le systÃ¨me marque ce point comme suspect.

**Explication de la sortie :**
Dans la cellule ci-dessous, nous injectons volontairement un prix de 1500 XOF/kg (trÃ¨s ÃŠlevÃŠ pour le maÃ¯s). 
Le tableau final n'affiche que cette ligne avec une nouvelle colonne `is_anomaly = True`. Le systÃ¨me a compris de lui-mÃĒme que 1500 FCFA/kg pour du maÃ¯s est totalement incohÃŠrent vis-Ã -vis des autres prix. Le module d'Aide Ã  la DÃŠcision ou de PrÃŠvision ignorera cette ligne biaisÃŠe grÃĸce Ã  ce tag.

```python
# Ajout d'une anomalie ÃŠvidente Ã  la fin de notre DataFrame (1500 FCFA/kg)
df_prices.loc[30] = {'date': pd.Timestamp.today(), 'price': 1500.0, 'unit': 'XOF/kg'}

# Application de l'algorithme de dÃŠtection
df_clean = pricing.detect_anomalies(df_prices)

# Affichage uniquement des anomalies trouvÃŠes
display(df_clean[df_clean['is_anomaly'] == True])
```
